```python
import pandas as pd
from pathlib import Path
```

```python
# import CSV
df = pd.read_csv('output.csv')
df.head()
```

```python
# start filtering, first just include music after 1930
since_1930 = df.loc[df['year'] > 1930]
since_1930 = since_1930[~since_1930['artist'].str.contains('genius', case=False)]
```

```python
# now, filter out artists with less than 20 songs
artist_appearances = since_1930['artist'].value_counts()
artists_with_more_than_20_songs = artist_appearances[artist_appearances > 20]
new_df = since_1930[since_1930['artist'].isin(artists_with_more_than_20_songs.index)]
new_df
```

```python
# and now by views
high_views_df = new_df[new_df['views'] >= 500000]
high_views_df
```

```python
# how many times does a specific artist show up in the data? 
kanye_west_appearances = (high_views_df['artist'] == 'Kanye West').sum()
kanye_west_appearances
```

```python
# let's see the most positive people
top_10_positive = high_views_df.sort_values(by='Positive', ascending=False).head(10)
top_10_positive
```
